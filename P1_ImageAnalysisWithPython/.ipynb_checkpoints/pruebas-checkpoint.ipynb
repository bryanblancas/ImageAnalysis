```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image

def colorToGrayscale(image_array):
    
    height = image_array.shape[0]
    width = image_array.shape[1]
    
    # new array with only 1 channels due to grayscale, without alpha
    gs_img_arr = np.zeros((height, width),dtype=np.int0)
    
    for i in range(0, height):
        for j in range(0, width):
            G =  int( 0.2126 * float( image_array[i,j,0] )) #redChannel 0
            G += int( 0.7152 * float( image_array[i,j,1] )) #greenChannel 1
            G += int( 0.0722 * float( image_array[i,j,2] )) #blueChannel 2
            
            gs_img_arr[i,j] = G
            
    return gs_img_arr



def histogramOfGrayscale(gs_img_arr):
    height = image_array.shape[0]
    width = image_array.shape[1]
    hist = np.zeros(255, dtype = np.int0);

    for i in range(0, height):
        for j in range(0, width):
            hist[gs_img_arr[i,j]] += 1
    
    return hist


def valueOfW(begin, end, image_histogram):
    w = 0.0
    for i in range(begin, end+1):
        w += image_histogram[i]
    return w

def valueOfM(begin, end, denominator, image_histogram):
    m = 0.0
    for i in range(begin, end+1):
        m += i * (image_histogram[i]/denominator)
    return m

def thresholding(gs_img_arr, T):
    arr = gs_img_arr.copy()
    height = gs_img_arr.shape[0]
    width = gs_img_arr.shape[1]
    
    for i in range(height):
        for j in range(width):
            if( arr[i,j] < T):
                arr[i,j] = 0
            else:
                arr[i,j] = 255
                
    return arr
        
def otsuMethod(denominator, image_histogram):
    T = 0
    rmax = 0

    for t in range(255):

        w0 = valueOfW(0, t, image_histogram) / denominator
        w1 = valueOfW(t+1, 254, image_histogram) / denominator
        print(str(w0) + " + " + str(w1) + " : " + str(w0+w1)) 

        m0 = valueOfM(0, t, denominator, image_histogram) / w0
        if w1 != 0:
            m1 = valueOfM(t+1, 254, denominator, image_histogram) / w1
        else:
            m1 = 9999999999
        #print(str(m0) + " + " + str(m1))

        r = w0 * w1 * ((m0-m1)**2)
        #print(r)

        if(r > rmax):
            rmax = r
            T = t
    
    return T

### Open an image
image_name = "./coins.png"
image = Image.open(image_name)
# type(image)
# image

### Convert image to numpy array
image_array = np.asarray(image)
print(image_array.shape)
# plt.imshow(image_array)



####### Convert image to grayscale
gs_img_arr = colorToGrayscale(image_array)
# print("gs_img_arr: ") 
# gs_img_arr
#plt.imshow(gs_img_arr, cmap = 'gray')




###### Convert grayscale image to binary image
image_histogram = histogramOfGrayscale(gs_img_arr)
plt.plot(image_histogram, color = "b")
plt.show()

##OTSU ALGORITM
denominator =  image_array.shape[0] * image_array.shape[1]
T = otsuMethod(denominator, image_histogram)

#print(T)

binary_image_arr = thresholding(gs_img_arr, T)
plt.imshow(binary_image_arr, cmap = 'gray')
    
```

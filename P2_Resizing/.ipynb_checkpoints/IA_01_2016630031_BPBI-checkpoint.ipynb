# **Image resizing**

Image interpolation occurs when you resize or distort your image from one pixel grid to another. Image resizing is necessary when you need to increase or decrease the total number of pixels, whereas remapping can occur when you are correcting for lens distortion or rotating an image. Zooming refers to increase the quantity of pixels, so that when you zoom an image, you will see more detail.

Interpolation works by using known data to estimate values at unknown points. For this practice, I will use **linear interpolation**.

### **Linear Interpolation**

It is a method to construct new data points within the range of a discrete set of known data points. Since this is a linear method, we use an straight line between two points previusly known. ### **Resizing process**

#### *Sampling*

The algorithm starts with the sampling of an image, i.e. it's necessary to obtain the values of the image in order to make the process works. It's necessary to say that any digital image lost information in the sampling process, this is because computers are not able to have a continuos function. In computers we have always discrete functions.

In the cells bellow, I show you the implementation of resizing algorithm in python and optimized with cython. Each cell has an explanation of his behaviour.

**Load Cython extension**

```python
# Load cython
%load_ext Cython
```

**Implementation of nearest integer function**

Due to the nature of the problem, this functions only works for positive values

```python
%%cython -a
# Only works for positive values
def nearest_integer(float i):
    if( float(i-int(i)) >= 0.5 ):
        return int(i)+1
    return int(i)
```

**Implementation of straight equation**

Due to the nature of the problem, x2 - x1 is always equal to 1

```python
%%cython -a
# y = (x-x1)( (y2-y1) / (x2-x1) ) + y1
# For this algorithm x2-x1 = 1 always, so:
# y = (x-x1)(y2-y1) + y1
def line(float x, int x1, int y1, int x2, int y2):
    cdef float f1 = x-x1
    cdef int f2 = int(y2)-int(y1)
    cdef float y = ( float(f1) * float(f2) ) + float(y1)
    return y
```

**Auxiliar function to calculate the new ratio of the resizing**

```python
%%cython -a
# Calculate ratio given by:
# original_size:x -> new_size:x1
# variable_a      ->        1
def calculate_ratio(float x, int x1):
    return x / float(x1)
```

**Implementation of the resizing method**

The code contains comments for better understanding

```python
%%cython -a
from __main__ import calculate_ratio
from __main__ import line
from __main__ import nearest_integer
cimport numpy as np

def resize_image(np.ndarray[np.npy_uint8, ndim=3] image_array, int new_height, int new_width):

    # Get original size of image
    cdef int height = image_array.shape[0]
    cdef int width = image_array.shape[1]
    cdef int channels = image_array.shape[2]

    # Create a copy of the image
    cdef np.ndarray resized_image = image_array.copy()

    #####################
    #	FOR NEW HEIGHT  #
    #####################

    # Resize the image with the new value of height
    resized_image.resize(new_height, width, channels)

    # Calculate new ratio
    cdef a = calculate_ratio(height, new_height)
    cdef char current_channel = 0
    cdef int current_column = 0
    cdef int current_row = 0
    cdef int i = 1
    cdef float x = 0
    cdef int x1 = 0
    cdef int y1 = 0
    cdef int x2 = 0
    cdef int y2 = 0
    cdef int new_value = 0
    
    # Traversal all channels, the whole width and the new height
    for current_channel in range(3):
        for current_column in range(width):
            for i in range(1, new_height):

                # Current value of new pixel
                x = i * a

                # Values of nearest pixel to the left of X
                x1 = int(x)
                y1 = image_array[x1, current_column, current_channel]

                # Values of nearest pixel to the right of X
                x2 = x1 + 1
                if(x2 == height):
                    x2 -= 1
                y2 = image_array[x2, current_column, current_channel]

                # Calculate new tonal value of new pixel i
                new_value = nearest_integer(line(x, x1, y1, x2, y2))

                # Assigment of new tonal value in the resized image
                resized_image[i, current_column, current_channel] = new_value

    #####################
    #	FOR NEW WIDTH   #
    #####################

    # Create a new array in order of set the new width
    # Now resized_image, i.e. the image with a new height is our initial image
    cdef np.ndarray resized_image2 = resized_image.copy()
    resized_image2.resize(new_height, new_width, channels)

    # Calculate new ratio
    a = calculate_ratio(width, new_width)

    #Traversal all channels, the whole width and the new width
    for current_channel in range(3):
        for current_row in range(new_height):
            for i in range(1, new_width):

                # Current value of new pixel
                x = i * a

                 # Values of nearest pixel to the left of X
                x1 = int(x)
                y1 = resized_image[current_row, x1, current_channel]

                # Values of nearest pixel to the right of X
                x2 = x1 + 1
                if(x2 == width):
                    x2 -= 1
                y2 = resized_image[current_row, x2, current_channel]

                 # Calculate new tonal value of new pixel i
                new_value = nearest_integer(line(x, x1, y1, x2, y2))

                # Assigment of new tonal value in the resized image
                resized_image2[current_row, i, current_channel] = new_value

    # Return resized image
    return resized_image2
```

# **Rest of the code**

Now, the rest of the code is shown in order. This section contains the integration of all the functions above as well as the libraries used. Also, the code contains the code lines necessary to show the process and the final result of the algorithm, i.e, theresized-image when the notebook is executed.

**1. Import <a href="https://numpy.org">NumPy</a>, <a href="https://matplotlib.org">matplotlib</a> and Image from <a href="https://pillow.readthedocs.io/en/stable/">PIL</a>**

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

**2. Open an image**

```python
image_name = "./guitar.jpg"
image = Image.open(image_name)
image_array = np.asarray(image)
```

**3. Asking for the new width and height of the image**

```python
print("Original size: "+str(image_array.shape))
new_height = int(input("New height: "))
new_width = int(input("New width: "))
```

**4. Executing the resize image function**

```python
%time resized_image = resize_image(image_array, new_height, new_width)
```

**4. Showing the original image**

```python
print("Original size: " + str(image_array.shape))
plt.imshow(image_array)
```

**5. Showing the resized image**

```python
print("New size: " + str(resized_image.shape))
plt.imshow(resized_image)
```
